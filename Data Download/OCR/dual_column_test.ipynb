```python
import io
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfpage import PDFPage

# Function to extract text from a PDF file using PDFMiner
def extract_text_from_pdf(pdf_path):
    resource_manager = PDFResourceManager()
    output = io.StringIO()
    codec = "utf-8"
    laparams = LAParams()
    device = TextConverter(resource_manager, output, codec=codec, laparams=laparams)
    
    with open(pdf_path, "rb") as pdf_file:
        interpreter = PDFPageInterpreter(resource_manager, device)
        for page in PDFPage.get_pages(pdf_file):
            interpreter.process_page(page)
    
    text = output.getvalue()
    output.close()
    return text

# Function to process extracted text and identify nested tables
def process_text(text):
    # Custom parsing logic here to identify and structure data
    # This logic should include identifying the dual-column table and nested tables,
    # and organizing the data accordingly.
    # For simplicity, let's assume you're looking for specific keywords or patterns.
    
    dual_column_table_data = []
    nested_table_data = []
    
    lines = text.split("\n")
    in_dual_column_table = False
    in_nested_table = False
    
    for line in lines:
        # Detect the start of the dual-column table
        if "Basic Information" in line:
            in_dual_column_table = True
            continue
        # Detect the end of the dual-column table
        if "Dual Column Table End" in line:
            in_dual_column_table = False
            continue
        # Detect the start of a nested table
        if "Lab Results" in line:
            in_nested_table = True
            continue
        # Detect the end of a nested table
        if "Nested Table End" in line:
            in_nested_table = False
            continue
        
        if in_dual_column_table:
            dual_column_table_data.append(line)
        
        if in_nested_table:
            nested_table_data.append(line)
    
    # Return the extracted data
    return {
        "dual_column_table_data": dual_column_table_data,
        "nested_table_data": nested_table_data
    }

# Path to the PDF file
pdf_path = "your_pdf_with_nested_tables.pdf"

# Extract text from the PDF
pdf_text = extract_text_from_pdf(pdf_path)

# Process the extracted text to identify and structure data
result = process_text(pdf_text)

# Access the extracted data
dual_column_table_data = result["dual_column_table_data"]
nested_table_data = result["nested_table_data"]

# Print or further process the extracted data as needed
print("Dual Column Table Data:")
for line in dual_column_table_data:
    print(line)

print("\nNested Table Data:")
for line in nested_table_data:
    print(line)
```

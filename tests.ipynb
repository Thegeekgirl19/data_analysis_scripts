```python
def emsugg_filtered(df1):
    df1['EANDM_Suggested_code'] = None
    for ind, row in df1.iterrows():
        if len(row["EANDM_Accepted"]) == 2:
            accepted = ""
        else:
            accepted = row["EANDM_Accepted"][1:-1]
            
        if len(row["EANDM_Rejected"]) == 2:
            rejected = ""
        else:
            rejected = row["EANDM_Rejected"][1:-1]

            
        if len(accepted) == 0:
            li = []
        else:
            li = ast.literal_eval(accepted)
            li = [li]

        if len(rejected) == 0:
            li2 = []
        else:
            li2 = ast.literal_eval(rejected)
            li2 = [li2]

        li3 = list(li) + list(li2)
        # print(li3)
        # row["EANDM_Accepted"] = [str(x) for x in row["EANDM_Accepted"]]
        # row["EANDM_Rejected"] = [str(x) for x in row["EANDM_Rejected"]]
        # df1=df1.astype({"EANDM_Accepted":"str"})
        # df1=df1.astype({"EANDM_Rejected":"str"})
        
        # li = ast.literal_eval(row["EANDM_Accepted"])
        # li2 = ast.literal_eval(row["EANDM_Rejected"])
        
        

        li3 = li + li2
        if 99281 in li3:
            li3.remove(99281)
            
        # output = [str(x) for x in li3]



        df1.at[ind,'EANDM_Suggested_code'] = li3
        
```

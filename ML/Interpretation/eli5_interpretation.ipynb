```python
def highlight_same(data):
    
    highlighter = pd.DataFrame("", index=data.index, columns=data.columns)
    for col in data.columns[:-1]:
        for index in data.index[:-1]:
            
            if col == index:
                highlighter.loc[index, col] = 'background-color : springgreen'
                
            elif str(col)>str(index) and data.loc[index, col]>100:
                highlighter.loc[index, col] = 'background-color : lightskyblue'

            elif data.loc[index, col]>100:
                highlighter.loc[index, col] = 'background-color : PaleVioletRed'

    return highlighter
# crosstab = cs.style.apply(highlight_same, axis=None)
```

```python
import pickle
import eli5
from sklearn.feature_extraction.text import TfidfVectorizer
from eli5.sklearn import PermutationImportance

# Load the trained logistic regression model
with open('/home/trishalaswain/Documents/machine_learning/LR.pkl', 'rb') as model_file:
    model = pickle.load(model_file)

# Load the vectorizer
with open('/home/trishalaswain/Documents/machine_learning/tfidf.pkl', 'rb') as vectorizer_file:
    vectorizer = pickle.load(vectorizer_file)

# Define some example data
# data = ["This is an example sentence.",
#         "Another example for interpretation."]

# Transform the data using the vectorizer
X = vectorizer.transform(df_mismatch['clean_sen_MDM_SENTS'][0:5])

X_dense = X.toarray()

# Predict the model's output for the given data
predictions = model.predict(X)

# Get the feature names from the vectorizer
feature_names = vectorizer.get_feature_names_out()

positive_color = (120, 80, 50)  # Green
negative_color = (0, 80, 50)    # Red
perm = PermutationImportance(model, random_state=42).fit(X_dense, df_mismatch['EANDM_Final_code_1'][0:5])

# Create a variable to store highlighted text
highlighted_text = []

# Interpret each prediction using eli5
for text, x, prediction, acc, EANDM_Final_code_1 in zip(df_mismatch['clean_sen_MDM_SENTS'][0:10], X_dense, predictions, df_mismatch['account_number'], df_mismatch['EANDM_Final_code_1']):
    print(f"Account Number: {acc}")
    # print(f"Account Number: {acc}")
    # Explain the prediction as HTML
    expl = eli5.explain_prediction_sklearn(model, text, vec=vectorizer, feature_names=feature_names)
    html_explanation = eli5.formatters.html.format_as_html(expl)
    
    # Append the HTML-formatted explanation to the list
    acc = "Account Number:\t" + str(acc) + "\n"
    final = "Final_code:\t" + str(EANDM_Final_code_1) + "\n"
    pred = "Model Prediction:\t" + str(prediction) + "\n"
    highlighted_text.append(acc)
    highlighted_text.append(final)
    highlighted_text.append(pred)
    highlighted_text.append(html_explanation)

# Save the highlighted text to an HTML file
with open('highlighted_text.html', 'w', encoding='utf-8') as text_file:
    
    text_file.write('\n'.join(highlighted_text))

print("Highlighted text with colors saved to 'highlighted_text.html'.")
```

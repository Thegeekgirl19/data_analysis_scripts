```python
import pandas as pd
from sqlalchemy import create_engine
import regex as re
import time
from datetime import datetime
import os
from bs4 import BeautifulSoup
```

```python
source_folder = r"/home/trishalaswain/ClientData/SCP/SCP_JULY_HALF/docs_1198_JULY/1198/"
```

```python
df = pd.DataFrame()
```

```python
df['account_number'] = None
df['final_code'] = None
df['text'] = None
```

```python
count = 0
for sub1 in os.listdir(source_folder):
    for file in os.listdir(source_folder + '/'+sub1 + "/"):
        if file.endswith('pcs.xml'):
            # print(file)
            
            account_number = file.split('_')[0]
            df.loc[count,'account_number'] = account_number
            final_code = file.split('_')[3]
            print(final_code)
            df.loc[count,'final_code'] = final_code
            text = ''
            # print(final_code)
            path = str(source_folder +sub1+"/"+ file )
            fd = open(path, 'r')
            xml_file = fd.read()
            soup = BeautifulSoup(xml_file, 'lxml')
            for tag in soup.findAll("sentence"):
                text += tag["senttext"]
            df.loc[count,'text'] = text
            count += 1
```

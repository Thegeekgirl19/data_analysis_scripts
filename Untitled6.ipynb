```python
import pandas as pd
import ast
```

```python
df = pd.read_excel("/home/trishalaswain/Downloads/Data/SCP-5365-GREENVILLE COMMUNITY HOSPITAL EAST-2023-01-01-2023-01-31.xlsx", index_col=False)
```

```python
df1 = df[["accountNumber","EANDM_Final","EANDM_Suggested","EANDM_Accepted","EANDM_Rejected","EANDM_Added","EANDM_Accepted_Count","EANDM_Rejected_Count","EANDM_Added_Count"]]
```

```python
def clean_suggested(df1):
    df1['clean_suggested'] = None
    for ind, row in df1.iterrows():
        if len(row["EANDM_Accepted"]) == 2:
            accepted = ""
        else:
            accepted = row["EANDM_Accepted"][1:-1]
            
        if len(row["EANDM_Rejected"]) == 2:
            rejected = ""
        else:
            rejected = row["EANDM_Rejected"][1:-1]

            
        if len(accepted) == 0:
            li = []
        else:
            li = ast.literal_eval(accepted)
            li = [li]

        if len(rejected) == 0:
            li2 = []
        else:
            li2 = ast.literal_eval(rejected)
            li2 = [li2]

        li3 = li + li2
        df1.at[ind,'clean_suggested'] = li3
```

```python
clean_suggested(df1)
```

```python
import pandas as pd
import re
```

```python
import os

# folder path
dir_path = r'/home/trishalaswain/Downloads/PH_01_to_31_Jan_tsv/'

# list to store files
res = []

# Iterate directory
for path in os.listdir(dir_path):
    # check if current path is a file
    if os.path.isfile(os.path.join(dir_path, path)) and str(path).endswith(".tsv") and not str(path).endswith("HCPCS.tsv") and not str(path).endswith("CPT.tsv") and not str(path).endswith("CM.tsv")  and not str(path).endswith("EANDM.tsv"):
        print(path)
        res.append(dir_path + path)
```

```python
i = 0
l = []

for file in res:
    
    # filename = re.findall("SCP-\d.*", file)
    filename = re.findall("PH-.*", file)

    # print(str(filename))
    df = pd.read_table(file, index_col = False)
    # df['serviceDate'] = pd.to_datetime( df.serviceDate)
    # df = df[df.serviceDate>='2023-01-01']
    # # df = df.reset_index()
    # df = df[if '99281' in df.EANDM_Final]
    
    # concatenating df1 and df2 along rows
    l.append(df)
    
df_final = pd.concat(l)
```

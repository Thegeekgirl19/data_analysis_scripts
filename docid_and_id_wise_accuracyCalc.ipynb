```python
df['clean_suggested'] = None
def clean_suggested(x):
    suggested = str(x['new_suggested']).strip()
    # print('suggested',suggested)
    suggested_list = suggested.replace(" ", ", ")
    # print('suggested_list',suggested_list)
    return suggested_list
df['clean_suggested'] = df.apply(clean_suggested, axis=1)
```

```python
def get_counts(x):
    
    suggested = str(x['clean_suggested'])
    # suggested = str(x['old_suggested'])
    final = str(x['final_code'])

    suggested_set = set(suggested.split(', '))
    if '{No Code}' in suggested_set:
        # print(suggested_set)
        suggested_set.remove('{No Code}')
    final_set = set(final.split(', '))
    if '{}' in final_set:
        # print(suggested_set)
        final_set.remove('{}')
    
    accepted = len(suggested_set.intersection(final_set))
    rejected = len(suggested_set - final_set)
    added = len(final_set - suggested_set)
    # print(x)
    # print('accepted',suggested_set.intersection(final_set))
    # print('rejected',suggested_set - final_set)
    # print('added',final_set - suggested_set)
    # print('suggested_set',suggested_set)
    # print('final_set',final_set)
    # print('------------')
    return pd.Series({'accepted':accepted, 'rejected':rejected, 'added':added})

count_df =  df.apply(get_counts, axis=1)
```

```python
count_df
```

```python
count_df.accepted.sum(), count_df.rejected.sum(), count_df.added.sum()
```

```python
precision = count_df.accepted.sum() / (count_df.accepted.sum() + count_df.rejected.sum())
```

```python
recall = count_df.accepted.sum() / (count_df.accepted.sum() + count_df.added.sum())
```

```python
f1_score = (2 * precision * recall) / (recall + precision)
```

```python
df = pd.read_excel('/home/trishalaswain/Downloads/SCP-1240-MOUNT SINAI MEDICAL CENTER-2023-04-01-2023-04-30.xlsx', index_col=False)
```

```python
cols_to_use = ['processingDate','accountNumber', 'EANDM_Suggested', 'EANDM_Final', 'EANDM_Accepted', 'EANDM_Rejected', 'EANDM_Added', 'EANDM_Accepted_Count', 'EANDM_Rejected_Count', 'EANDM_Added_Count',];
```

```python
df['EANDM_Suggested_Code'] = None
df['EANDM_Final_Code'] = None
```

```python
def clean(df):
    for i,r in df.iterrows():
        df.at[i,'EANDM_Suggested_Code'] = ','.join([str(elem) for elem in re.findall(r"\b\d{5}\b", str(r['EANDM_Suggested']))])
        df.at[i,'EANDM_Final_Code'] = ','.join([str(elem) for elem in re.findall(r"\b\d{5}\b", str(r['EANDM_Final']))])
        return df[cols_to_use].fillna('')
```

```python
def get_counts(x):
    
    suggested = str(x['EANDM_Suggested_Code'])
    final = str(x['EANDM_Final_Code'])
    
    suggested_set = set(suggested.split(','))
    final_set = set(final.split(','))
    
    # if '9' not in final:
    #     # print(x)
    #     # print('final',final)
    #     # print('final_set', final_set)
    #     final_set.clear()
    #     # print('final_set afer clear: ', final_set)
        
    # if '9' not in suggested:
    #     # print(x)
    #     # print('suggested: ',suggested)
    #     # print('suggested_set: ', suggested_set)
    #     suggested_set.clear()
    #     # print('suggested_set after clear: ', suggested_set)
        
    
    accepted = len(suggested_set.intersection(final_set))
    rejected = len(suggested_set - final_set)
    added = len(final_set - suggested_set)
    
    if (rejected >= 1 or added >= 1):
        print('rejected',rejected)
        print('added',added)
        print('suggested_set',suggested_set,'\t final_set',final_set)
            
    return pd.Series({'accepted':accepted, 'rejected':rejected, 'added':added})
```

```python
df = df[cols_to_use]
clean_df = clean(df)
clean_df.dropna(inplace=True)
count_df =  clean_df.apply(get_counts, axis = 1)

# old_sum = df.EANDM_Accepted_Count.sum() + df.EANDM_Added_Count.sum()
# print(old_sum)
new_sum = count_df.accepted.sum() + count_df.added.sum()
# print("New Accepted + Added\t" +  str(new_sum))

precision = count_df.accepted.sum() / (count_df.accepted.sum() + count_df.rejected.sum())
recall = count_df.accepted.sum() / (count_df.accepted.sum() + count_df.added.sum())
f1_score = (2 * precision * recall) / (recall + precision)
print(precision, recall, f1_score)
```

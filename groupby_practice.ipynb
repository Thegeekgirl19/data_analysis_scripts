```python
df2 = pd.DataFrame({"power_level": [12000, 16000, 4000, 1500, 3000, 
                                     2000, 1600, 2000, 300],
                     "uniform color": ["orange", "blue", "black", "orange",
                                       "purple", "green", "orange", "orange","orange"],
                     "species": ["saiyan","saiyan","saiyan","half saiyan",
                                 "namak","human","human","human","human"]}, 
                     index = ["Goku","Vegeta", "Nappa","Gohan",
                                   "Piccolo","Tien","Yamcha", "Krillin","Roshi"])
```

```python
df2
```

```python
def mode(x):
    return x.mode()
df2.apply(mode, axis = 0)
```

```python
def power_stats(x, h):
    if x > h:
        return 'high'
    if x == h:
        return 'equal'
    if x < h:
        return 'low'
df2['level_stats'] = df2['power_level'].apply(power_stats, args=[4000])
```

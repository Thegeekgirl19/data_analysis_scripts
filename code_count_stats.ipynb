```python
import pandas as pd
```

```python
def initialize(index):
    df_final.at[index,'1_doc_only_counts'] = 0
    df_final.at[index,'2_doc_only_counts'] = 0
    df_final.at[index,'3_doc_only_counts'] = 0
    df_final.at[index,'4_doc_only_counts'] = 0
    df_final.at[index,'other_no_of_docs_counts'] = 0
```

```python
import os

dir_path = r'/home/trishalaswain/Downloads/PIMG-STATS/'

res = []

for path in os.listdir(dir_path):
#     if path.startswith("."):
#         continue
#     if '28.tsv' not in path:
#         continue
#     else:
    res.append(path)
```

```python
def fill_df_final(index,facility_code,facility_name,total_no_of_encounters,total_no_of_docs):
    df_final.at[index,'facility_code'] = str(facility_code)
    df_final.at[index,'facility_name'] = facility_name
    df_final.at[index,'no_of_encounters'] = total_no_of_encounters
    df_final.at[index,'total_docs'] = total_no_of_docs
    initialize(index)
```

```python
from collections import defaultdict
```

```python
def get_counts(code_list, df_to_use, col, codetype):
    count_dict = {}
    for code in code_list:
        count_dict[code+'_'+codetype] = df_to_use[col].str.contains(code).sum()
        
    return count_dict
```

```python
code_list = ['99281','99282','99283','99284','99285', '99291', '99292','99239','99238','99232','99233','99497','99223','99221','99231']
```

```python
l = []

for item in res:
#     if '2563' not in item:
#         continue
    
    path = '/home/trishalaswain/Downloads/PIMG-STATS/'
#     df = pd.read_csv(path+item,sep='\t',index_col = False)
    df = pd.read_excel(path+item,index_col = False)
    
#     print(df.shape)
#     print(item)
#     if str(df["month"]).__contains__("22-12"):
#         continue

#     df = df[df.month=='2023-01']

#     df['serviceDate'] = pd.to_datetime( df.serviceDate)
    
#     df = df[df.serviceDate>='2023-01-01']
    
#     print(df['EANDM_Suggested'])
#     print(df.shape)
    if df.shape[0]<1:
        continue
    
#     code_list = ['99281','99282','99283','99284','99285']
    
#     for i in code_list:
#         find_count_of_code(i,df,"final")
#         find_count_of_code(i,df,"suggested")
    
#     print(get_counts(code_list, df, 'EANDM_Suggested', 'Suggested')) 
#     print(get_counts(code_list, df, 'EANDM_Final', 'Final'))
    
    df['doc_count'] = df['encDocIds'].apply(lambda x:len(str(x).split(', ')))

    final_dict = {}
    
    final_dict['facility_code'] = df.iloc[0].loc['facilityCode']
    final_dict['facility_name'] = df.iloc[0].loc['facilityName']
    final_dict['no_of_encounters'] = df.shape[0]
    final_dict['total_docs'] = df.doc_count.sum()
    

    new_dict = defaultdict(int)
    for k, v in df.doc_count.value_counts().to_dict().items():
        if k >= 1 and k <= 4:
            new_dict["{}_doc_only_counts".format(k)] = v
        else:
            new_dict["other_no_of_docs_counts"]+= v
            
    final_dict['1_doc_only_counts'] = new_dict['1_doc_only_counts']
    final_dict['2_doc_only_counts'] = new_dict['2_doc_only_counts']
    final_dict['3_doc_only_counts'] = new_dict['3_doc_only_counts']
    final_dict['4_doc_only_counts'] = new_dict['4_doc_only_counts']
    final_dict['other_no_of_docs_counts'] = new_dict['other_no_of_docs_counts']


    final_dict.update(df['doc_count'].describe().to_dict())
    
#     display(df['EANDM_Accepted'].value_counts())
    
    final_dict.update(get_counts(code_list, df, 'EANDM_Suggested', 'Suggested'))
    final_dict.update(get_counts(code_list, df, 'EANDM_Final', 'Final'))
    
    l.append(final_dict)
#     print(item)
```

```python
final_analysis = pd.DataFrame(l)
```

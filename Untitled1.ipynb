```python
def get_counts(x):
    
    suggested = str(x['improved_suggested']).strip()
    # suggested = suggested.replace('[', '').replace(']', '').replace("'", '').replace("'", '')
    
    final = str(x['EANDM_Final']).strip()
    # print(final)
    # final = final.replace('[', '').replace(']', '').replace("'", '').replace("'", '')
    # print(final)


    suggested_set = set(suggested.split(', '))
    # if 'No valid sections detected' in suggested_set:
    #     suggested_set.remove('No valid sections detected')
    
    final_set = set(final.split(', '))
    # if '99291' in final_set:
    #     final_set.remove('99291')
    #     suggested_set.clear()
    # if '99292' in final_set:
    #     final_set.remove('99292')
    #     suggested_set.clear()
        
    # if '99291' in suggested_set:
    #     final_set.clear()
    # if '99292' in suggested_set:
    #     final_set.clear()
    
    accepted = len(suggested_set.intersection(final_set))
    rejected = len(suggested_set - final_set)
    added = len(final_set - suggested_set)
    if (rejected >= 1 or added >= 1):
        print(x['accountNumber'])
        print('rejected',rejected)
        print('added',added)
        print('suggested_set',suggested_set,'\t final_set',final_set)
            
    return pd.Series({'accepted':accepted, 'rejected':rejected, 'added':added})
```

```python
count_df =  df_use_2.apply(get_counts, axis=1)
```

```python
precision = count_df.accepted.sum() / (count_df.accepted.sum() + count_df.rejected.sum())
```

```python
recall = count_df.accepted.sum() / (count_df.accepted.sum() + count_df.added.sum())
```

```python
f1_score = (2 * precision * recall) / (recall + precision)
```
